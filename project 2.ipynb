# Storytelling Data Visualization on Exchange Rates


```python
import pandas as pd
import warnings

warnings.filterwarnings("ignore")
```

```python
exchange_rates = pd.read_csv("euro-daily-hist_1999_2022.csv")
exchange_rates
```

```python
exchange_rates.shape
```

```python
exchange_rates.info()
```

```python
exchange_rates.rename(columns={'[US dollar ]': 'US_dollar','Period\\Unit:':'Time'},inplace = True)
```

```python
exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])
```

```python
exchange_rates.sort_values('Time',inplace = True)
```

```python
exchange_rates.head()
```

```python
euro_to_dollar = exchange_rates[['Time','US_dollar']].copy()
euro_to_dollar['US_dollar'].value_counts()
```

```python
euro_to_dollar = euro_to_dollar[euro_to_dollar['US_dollar'] != '-']
euro_to_dollar['US_dollar'] = euro_to_dollar['US_dollar'].astype(float)
euro_to_dollar.info()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(euro_to_dollar['Time'],euro_to_dollar['US_dollar'])
plt.show()
```

```python
values = pd.DataFrame()
values['daily_values'] = pd.Series(range(1,20,2))
values
```

```python
values['rolling_mean_2'] = values['daily_values'].rolling(2).mean()
values
```

```python
values['rolling_mean_3'] = values['daily_values'].rolling(3).mean()
values['rolling_mean_5'] = values['daily_values'].rolling(5).mean()
values
```

```python
plt.figure(figsize = (9,6))

plt.subplot(3,2,1)
plt.plot(euro_to_dollar['Time'],euro_to_dollar['US_dollar'])
plt.title('Original values',weight = 'bold')

for i, rolling_mean in zip([2,3,4,5,6],
                            [7,30,50,100,365]):
    plt.subplot(3,2,i)
    plt.plot(euro_to_dollar['Time'],euro_to_dollar['US_dollar'].rolling(rolling_mean).mean())
    plt.title('Rolling Windows: ' + str(rolling_mean),weight = 'bold')
    plt.ylim(float(0.75),float(1.6))
    
plt.tight_layout()
plt.show()
```

```python
euro_to_dollar['rolling_mean'] = euro_to_dollar['US_dollar'].rolling(30).mean()
euro_to_dollar
```

```python

financial_crisis = euro_to_dollar.copy()[
    (euro_to_dollar['Time'].dt.year >= 2006) & (euro_to_dollar['Time'].dt.year <= 2009)
]

financial_crisis_7_8 = euro_to_dollar.copy()[
    (euro_to_dollar['Time'].dt.year >= 2007) & (euro_to_dollar['Time'].dt.year <= 2008)
]
```

```python
import matplotlib.style as style
style.use('fivethirtyeight')

fig,ax = plt.subplots(figsize= (8,3))
ax.plot(financial_crisis['Time'],
        financial_crisis['rolling_mean'],
        linewidth = 1, color = 'blue')

ax.plot(financial_crisis_7_8['Time'],
        financial_crisis_7_8['rolling_mean'],
        linewidth = 2, color = 'red')

ax.set_xticklabels([])

x = 0.02
for year in ['2006','2007','2008','2009','2010']:
    ax.text(x, -0.08, year, alpha = 0.6, fontsize = 12, transform = plt.gca().transAxes)
    x += 0.22888
    
ax.set_yticklabels([])

y = 0.07
for rate in ['1.2','1.3','1.4','1.5']:
    ax.text(-0.04, y, rate, alpha = 0.6, fontsize = 12, transform = plt.gca().transAxes)
    y += 0.2333
    
ax.axvspan(xmin = pd.to_datetime("2008-04-1"), xmax = pd.to_datetime("2008-09-1"), ymin = 0.09,
          alpha = 0.4, color = 'grey')

ax.text(-0.05,1.3, "Euro-USD rate peaked at 1.59 during 2007-2008's financial crisis", weight = 'bold',
       transform = plt.gca().transAxes)

ax.text(-0.05,1.2, "Euro-USD exchnage rates between 2006 and 2010",size = '12',
       transform = plt.gca().transAxes)

ax.text(-0.05, -0.25, 'Abdullah' + ' ' * 80 + 'Source: European Central Bank', color='#f0f0f0',
        backgroundcolor='#4d4d4d', size=13, transform=plt.gca().transAxes)

plt.show()
```

```python
corona_crisis_20 = euro_to_dollar.loc[(euro_to_dollar["Time"]>= '2020-01-01')&
                                  (euro_to_dollar['Time']<= '2020-12-31')]
corona_crisis = euro_to_dollar.loc[(euro_to_dollar['Time']>= '2016-01-01')&
                                  (euro_to_dollar['Time']<= '2019-12-31')]
```

```python
import matplotlib.style as style
style.use('fivethirtyeight')

fig,ax = plt.subplots(figsize= (9,3))
ax.plot(corona_crisis_20 ['Time'],
        corona_crisis_20 ['rolling_mean'],
        linewidth = 3, color = 'red')

ax.plot(corona_crisis['Time'],
        corona_crisis['rolling_mean'],
        linewidth = 1, color = 'blue')

ax.set_xticklabels([])

x = 0.02
for year in ['2016','2017','2018','2019','2020','2021']:
    ax.text(x, -0.08, year, alpha = 0.6, fontsize = 12, transform = plt.gca().transAxes)
    x += 0.183
    
ax.set_yticklabels([])

y = 0.07
for rate in ['1.05','1.10','1.15','1.20']:
    ax.text(-0.04, y, rate, alpha = 0.6, fontsize = 12, transform = plt.gca().transAxes)
    y += 0.248
    


ax.text(-0.05,1.3, "Euro-USD rate peaked at 1.22 during 2020's COVID19 crisis", weight = 'bold',
       transform = plt.gca().transAxes)

ax.text(-0.05,1.2, "Euro-USD exchnage rates between 2019 and 2020",size = '12',
       transform = plt.gca().transAxes)

ax.text(-0.05, -0.25, 'Abdullah' + ' ' * 80 + 'Source: European Central Bank', color='#f0f0f0',
        backgroundcolor='#4d4d4d', size=13, transform=plt.gca().transAxes)


plt.show()
```

```python
style.use('fivethirtyeight')

plt.figure(figsize= (14,8))

ax1 = plt.subplot(3,3,1)
ax2 = plt.subplot(3,3,2)
ax3 = plt.subplot(3,3,3)
ax4 = plt.subplot(3,1,2)

axes = [ax1,ax2,ax3,ax4]

for ax in axes:
    ax.set_ylim(0.8,1.7)
    ax.set_yticks([1.0,1.2,1.4,1.6])
    ax.set_yticklabels(['1.0','1.2','1.4','$1.6'],
                      alpha = 0.5)
    
ax1.plot(bush['Time'],bush['rolling_mean'],
        color = '#BF5FFF')
ax1.set_xticklabels(['','2001','','2003','','2005','','2007','','2009'],
                   alpha = 0.4, size = 11)
ax1.text(0.11,2.45,'BUSH',fontsize = 20, weight = 'bold', color = '#BF5FFF', transform = plt.gca().transAxes)

ax1.text(0.09,2.34,'(2001 - 2009)',weight = 'bold', color = '#BF5FFF', alpha = 0.4, transform = plt.gca().transAxes)

ax2.plot(obama['Time'],obama['rolling_mean'],
        color = '#ffa500')
ax2.set_xticklabels(['','2009','','2011','','2013','','2015','','2017'],
                   alpha = 0.4, size = 11)
ax2.text(0.45,2.45,'ObAMA',fontsize = 20, weight = 'bold', color = '#ffa500', transform = plt.gca().transAxes)

ax2.text(0.44,2.34,'(2009 - 2017)',weight = 'bold', color = '#ffa500', alpha = 0.4, transform = plt.gca().transAxes)

ax3.plot(trump['Time'],trump['rolling_mean'],
        color = '#00B2EE')
ax3.set_xticklabels(['','2017','','2018','','2019','','2020','','2021'],
                   alpha = 0.4, size = 11)
ax3.text(0.82,2.45,'TRUMP',fontsize = 20, weight = 'bold', color = '#00B2EE', transform = plt.gca().transAxes)

ax3.text(0.808,2.34,'(2017 - 2021)',weight = 'bold', color = '#00B2EE', alpha = 0.4, transform = plt.gca().transAxes)

ax4.plot(bush['Time'],bush['rolling_mean'],
        color = '#BF5FFF')
ax4.plot(obama['Time'],obama['rolling_mean'],
        color = '#ffa500')
ax4.plot(trump['Time'],trump['rolling_mean'],
        color = '#00B2EE')

ax4.set_xticks([])

ax.text(-0.05,2.8, "Euro-USD rate averaged 1.22 under the last three US presidents", weight = 'bold',
       transform = plt.gca().transAxes)

ax.text(-0.05,2.7, "Euro-USD exchnage rates under Geaorge W. Bush (2001 - 2009), Barack Obama (2009 - 2017) and Donald Trump (2017 - 2021)",size = '12',
       transform = plt.gca().transAxes)

ax4.text(-0.05, -0.15, 'Abdullah' + ' ' * 168+ 'Source: European Central Bank', color='#f0f0f0',
        backgroundcolor='#4d4d4d', size=13, transform=plt.gca().transAxes)
plt.tight_layout()
plt.show()

```
